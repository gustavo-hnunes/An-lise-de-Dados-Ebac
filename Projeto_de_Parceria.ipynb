```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('salary_data.csv', sep=',')
data.isnull().any()
```

```python
data
```

```python
median_array = np.asarray(data['median_salary'].astype('float').tolist())
average_array = np.asarray(data['average_salary'].astype('float').tolist())
lowest_array = np.asarray(data['lowest_salary'].astype('float').tolist())
highest_array = np.asarray(data['highest_salary'].astype('float').tolist())
```

```python
median_media = data['median_salary'].mean()
average_media = data['average_salary'].mean()
lowest_media = data['lowest_salary'].mean()
highest_media = data['highest_salary'].mean()

median_desvio = data['median_salary'].std()
average_desvio = data['average_salary'].std()
lowest_desvio = data['lowest_salary'].std()
highest_desvio = data['highest_salary'].std()

data['median_salary_std'] = data['median_salary'].apply(lambda salario_mediano: (salario_mediano - median_media) / median_desvio)
data['average_salary_std'] = data['average_salary'].apply(lambda salario_medio: (salario_medio - average_media) / average_desvio)
data['lowest_salary_std'] = data['lowest_salary'].apply(lambda salario_menor: (salario_menor - lowest_media) / lowest_desvio)
data['highest_salary_std'] = data['highest_salary'].apply(lambda salario_maior: (salario_maior - highest_media) / highest_desvio)

data.head()
```

```python
from os import spawnl
data['continent_name'].drop_duplicates()

continent_mapper = {
    "Asia": 'ASI',
    "Europe": 'EUR',
    "Africa": 'AFR',
    "Oceania": 'OCE',
    "Caribbean": 'CAR',
    "South America": 'SAM',
    "Central America": 'CAM',
    "North America": 'NAM',
    "Northern America": 'NAM'
}

data['continent_nom'] = data['continent_name'].apply(lambda continent: continent_mapper[continent])
data.head()

span_mapper = {
    "Weekly": 1,
    "Monthly": 2,
    "Semester": 3,
    "Yearly": 4
    }

data['wage_span_nom'] = data['wage_span'].apply(lambda span: span_mapper[span])
```

```python
data = data.drop(['continent_name', 'median_salary', 'lowest_salary', 'highest_salary', 'average_salary', 'wage_span'], axis=1)
data = data.reindex(columns=['country_name', 'continent_nom', 'wage_span_nom', 'median_salary_std', 'average_salary_std', 'lowest_salary_std', 'highest_salary_std'])
data.head()
```
